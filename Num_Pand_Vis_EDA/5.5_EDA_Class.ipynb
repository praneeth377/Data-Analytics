```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',30)
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.shape
```

```python
plt.hist(df['Distance'], bins=50)
```

```python
sns.distplot(df['Distance'], hist=False)
```

```python
df['Distance'].median()
```

```python
plt.figure(figsize=(20,10))
plt.hist(df['Distance'], bins=20, color='g')

plt.axvline(df['Distance'].mean(), linewidth=1)
plt.axvline(df['Distance'].median(), linestyle='dashed', linewidth=1)
plt.axvline(df['Distance'].mode()[0], linestyle='dotted', linewidth=1)

plt.xticks(ticks=[df['Distance'].median(), df['Distance'].mode()[0], df['Distance'].mean()], labels=[df['Distance'].median(), df['Distance'].mode()[0], df['Distance'].mean()], fontsize=6)
plt.show()
```

```python
sns.distplot(df['Distance'], hist_kws=dict(cumulative=True), kde_kws=dict(cumulative=True))
```
