```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
data.head()
data
```

```python
x = np.linspace(0,10,100)
y = np.exp(x)
plt.xlabel('X-Axis')
plt.ylabel('Y-axis')
plt.title('Exponential Graph')

plt.plot(x,y,'g*')
plt.show()
```

```python
x = np.linspace(-100,100,1000)          # Standard Logit Curve 
L = 1
k = 1
x_0 = 0
y = L/(1+(np.exp(-k*(x-x_0))))
plt.ylim(0,1)
plt.xlim(-5,5)
plt.plot(x,y,'g')
plt.show()
```

```python
x = np.linspace(-10,10,100)
y1 = x
y2 = x**2
y3 = np.sqrt(x)
y4 = x**3

plt.ylim(-5,5)
plt.axhline(color = 'black', linewidth=0.5)
plt.axvline(color = 'black', linewidth=0.5)

plt.plot(x,y1,'r',x,y2,'g',x,y3,'b',x,y4,'orange')
plt.show()
```

```python
x = data['total_bill']
y = data['tip']

plt.scatter(x,y,s=np.array(y)**2,c='green')   # custom sizing so that large tips will be large dots 

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot')
plt.show()
```

```python
# Bar Plot basic example
amount = [10000,21000,5000,8000,12500]
names = ['Arvind','Kesava','Chaitanya','Rahul','Mahesh']
#x_pos = np.arange(len(names))
x_pos = [1,2,3,4,5]
plt.bar(x_pos, amount, width= 0.6)
plt.xticks(x_pos, names)

plt.title('Distribution of Bill Amount')
plt.xlabel('Names')
plt.ylabel('Amount')
plt.figure(figsize=(20,10))

plt.show()
```

```python
d21 = pd.DataFrame({'Year':[2001,2002,2003,2004,2005], 'Import':[54.4,53.8,61.6,74.15,89.33], 'Export':[42.5,44.5,48.3,57.24,69.18]})


plt.bar(d21['Year'], height=d21['Import'], width=0.35, label='Import')
plt.bar(d21['Year']+0.35, height=d21['Export'], width=0.35, label='Export')

plt.xticks(ticks=d21['Year']+(0.35/2), labels=d21['Year'])
plt.title('Import and Export by Year')
plt.legend()
plt.show()
```

```python
plt.hist(data['tip'],bins=5,color='r',width=1.6,histtype='barstacked')
c,b = np.histogram(data['tip'], bins=5,)
print(c,b)
plt.xlabel('Bin Distribution')
plt.ylabel('Frequency')
plt.title('Tip Histogram')
plt.show()
```

```python
plt.boxplot(data['total_bill'])
plt.text(x=1.1,y=data['total_bill'].min(),s= data['total_bill'].min())
plt.text(x=1.1,y=data['total_bill'].max(),s= data['total_bill'].max())
plt.text(x=1.1,y=data['total_bill'].median(),s= f"Median: {data['total_bill'].median():.2f}")
plt.text(x=1.1,y=data['total_bill'].quantile(0.75),s= f"Q3: {data['total_bill'].quantile(0.75):.2f}")    # Good example of using f string and rounding off
plt.text(x=1.1,y=data['total_bill'].quantile(0.25),s= data['total_bill'].quantile(0.25))


plt.show()
```
