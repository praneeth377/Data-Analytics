### 5. Count the number of data values that fall within two standard deviations of the mean. Compare this with the answer from Chebyshev’s Theorem.

```python
datapoints_between_2stds = 0
for i in data:
    if data.mean()-2*data.std() <= i <= data.mean()+2*data.std():
        datapoints_between_2stds += 1
```

```python
print('No. of data values within 2 Standard Deviations is',datapoints_between_2stds)
print('% points within 2 Standard Deviations is',100*datapoints_between_2stds/len(data))
```

Mathematically, Chebyshev's Theorem is as follows

$P(\mu−k\sigma≤X≤\mu+k\sigma)≥1− 1/(k^2)$

That means, within 2 standard deviations there should be atleast 75% of data points. Data provided satisfies this theorem.

### 13. What is the shape of the distribution of this dataset? Create an appropriate graph to determine that. Take 100 random samples with replacement from this dataset of size 5 each. Create a sampling distribution of the mean shirt prices. Compare with other sampling distributions of sample size 10, 15, 20, 25, 30. State your observations. Does it corroborate the Central Limit Theorem?

As seen above, the shape nearly resembles normal distribution with slight right tailed-ness i.e., positive skewness.

Performing CLT on data:

Sample Size n = 5. Number of iterations = 100.

```python
n,iter = 5,100
np.random.seed(1)
sample_means = []
for i in range(iter):
    samples = np.random.choice(data,n)
    sample_mean = np.mean(samples)
    sample_means.append(sample_mean)

sns.kdeplot(sample_means, fill=True, color='darkolivegreen')
plt.title(f"Distribution for Sample Size {n}")
plt.show()
```

Increasing the Sample Size. Same number of iterations

```python
fig,axes = plt.subplots(2,3,figsize=(15,11))
axes = axes.ravel()
np.random.seed(1)

for i,j in zip(range(6),[10,15,20,25,30,35]):
    ax = axes[i]
    sample_means = []
    for k in range(100):
        samples = np.random.choice(data,j)
        sample_mean = np.mean(samples)
        sample_means.append(sample_mean)

    sns.kdeplot(sample_means, fill=True, ax=ax, color='darkorchid')
    ax.set_title(f"Distribution for Sample Size {j}")
    ax.text(0.7,0.9,f"Skew:{round(pd.Series(sample_means).skew(),3)}", transform=ax.transAxes)

plt.tight_layout()
plt.show()
```

We can observe that as sample size increases, the skewness is reducing. That means the shape is slowly converting into normal curve. This indeed corroborates with Central Limit Theorem.

### 16. Compute a 95% Confidence Interval for the true price of a shirt in the NuCloth website  using appropriate distribution. (State reasons as to why did you use a z or t distribution)

```python
print('Sample Size is',len(data))
```

Some sources say that if sample size is greater than 30 then to forget about t-distribution and go with standard normal distribution i.e., z-distribution. But since our sample size is just over 30 we have decided to go with t-distribution only as we don't know population variance and t-distribution allows for more variance.

```python
sample_mean = data.mean()
sample_std = data.std(ddof=1)  # ddof=1 for sample standard deviation
df = len(data) - 1             # For a t-distribution, degrees of freedom (df) is equal to the sample size minus 1.
```

```python
alpha = 0.05
t_score = t.ppf(0.975, df)     # Two-tailed, alpha/2 = 0.025 for 95% Confidence Interval
print(t_score)
```

Margin of Error for 95% confidence $E_{0.025} = t_{0.025}\times s/\sqrt{n}$

$s$ is standard deviation of sample (not for population).

```python
moe = t_score*(sample_std/np.sqrt(len(data)))
print(moe)
```

Confidence Interval would be $\bar{X} \pm E_{0.025}$

```python
confidence_interval = (sample_mean - moe, sample_mean + moe)
print('95% Confidence Interval for Mean Price of Shirt is',tuple(round(i,2) for i in confidence_interval))
```

### 17. A data scientist wants to estimate with 95% confidence the true proportion of shirts having price greater than 25 dollars in the NuCloth website.  A recent study showed that 60% of all shirts have a price greater than 25 dollars. The data scientist wants to be accurate within 2% of the true proportion. Find the minimum sample size necessary

The margin of error in the context of a confidence interval for a population proportion is given by:

$E = Z\times\sqrt{(p\times(1-p))/n}$

$\to$ $Z$ is the Z-score corresponding to the desired confidence level

$\to$ $p$ is the estimated population proportion

$\to$ $n$ is the sample size

Here we deal with normal distribution (Z-distribution) as we are dealing with proportions and the sample size is expected to be large enough for the Central Limit Theorem to apply.

Rearranging the formula to solve for sample size ($n$), we get

$n = (Z^2\times p\times (1-p))/E^2$

```python
E,p = 0.02,0.6
Z = norm.ppf(0.95 + (0.05/2))     # alpha/2 = 0.025 for 95% confidence interval
n = ((Z**2)*p*(1-p))/E**2
print('Minimum Sample Size:',round(n))
```

---

### 18. The same data scientist wants to estimate the true proportion of shirts having price greater than 25 dollars. She wants to be 90% confident and accurate within 5% of true proportion. Find the minimum sample size necessary

```python
# Same as above, just changing alpha and margin of error E.
E,p = 0.05,0.6
Z = norm.ppf(0.90 + (0.1/2))     # alpha/2 = 0.05 for 90% confidence interval
n = ((Z**2)*p*(1-p))/E**2
print('Minimum Sample Size:',round(n))
```

### 19. NuCloth claims that currently 80% of all shirts have prices greater than 25 dollars . Test this claim with an alpha = 0.05 if out of a random sample of 30 shirts only 23 are having prices above 25 dollars

We will use One Proportion Z-test. Let $p$ be the proportion of shirts having prices greater than 25 dollars.

> $H_0 : p = 0.8$

> $H_a : p \neq 0.8$

Let's test whether the Z-test assumptions are satisfied or not
*   Binomally distributed population - Yes, either the price is greater than 25 dollars or less than or equal to 2 dollars.
*   Can the binomial distribution approximated to normal distribution - Yes. For binary data, CLT works slower than usual. The standard thing is to check whether np and n(1-p) are greater than or equal to 10. 
>$np = 30\cdot \frac{23}{30} =23 \geq 10\\
n(1-p) = 30 \cdot \frac{30-23}{30} =7 \leq 10$

Although not satisfied, we will proceed with the test. Ideally we can't proceed but in this case let's make an exception.

Test Statistic $Z = \sqrt{n}\times(\hat{p} - p)/\sqrt{p\times (1-p)}$

* $\hat{p}$ is the sample proportion
* $p$ is  the claimed population proportion
* $n$ is the sample size

The statsmodels library offers a convenient pre-built function, proportions_ztest, that automates the calculation of the test statistic and p-value for the z-test for proportions

```python
from statsmodels.stats.proportion import proportions_ztest

test_stat, p_value = proportions_ztest(23, 30, value = 0.8, alternative = 'two-sided')
print('Z-statistic is',test_stat)
print('P-value is',p_value)
```

As the p-value is much greater than the significance level 0.05, we can't reject the $H_0$. Hence the claim that 80% shirts have prices greater than 25 dollars is acceptable as there is enough statistical evidence supporting it.

### 20. A data scientist is researching the hypothesis that the average price of a shirt in NuCloth is higher than the supermarket. So he collects data from NuCloth and the supermarket that the average shirt price of shirts  in NuCloth is 36 dollars vs 34 dollars in the supermarket. The standard deviations are 7.6 and 5.8 respectively. Suppose the data scientist got these values after randomly selecting 100 shirts from each place

### a. What hypothesis would he use to compare the shirt prices of NuCloth vs Supermarket

Let $\mu_1$ be average shirt price in NuCloth and $\mu_2$ be average shirt price in supermarket. 

Null and Alternate Hypotheses :-

> $H_0 : \mu_1 = \mu_2$

> $H_a : \mu_1 > \mu_2$

### b. What are critical values to be used?

We chose the 95% confidence interval i.e., critical value with which we compare our P-value would be 0.05.

### c. What statistical test will be used to compare these prices?

We have two independent samples with known sample means and sample deviations. We don't know anything about the population. The ideal statistical test would be "Two Independent Sample T-test for Equality of Means - Unequal Standard Deviation"

### d. Complete the test and obtain the P-value. 

Test Statistic $t = (\bar{X_1} - \bar{X_2})/\sqrt{s_1^2/n_1 + s_2^2/n_2}$

* $\bar{X_1}$, $\bar{X_2}$ are the sample means for NuCloth and the supermarket, respectively.
* $s_1$, $s_2$ are the sample standard deviations for NuCloth and the supermarket, respectively.
* $n_1$, $n_2$ are the sample sizes for NuCloth and the supermarket, respectively.

Again, we make use of pre-built function below..

```python
from scipy.stats import ttest_ind_from_stats

test_stat, p_value = ttest_ind_from_stats(36,7.6,100,34,5.8,100, equal_var=False, alternative='greater')
print("t-statistic:", test_stat)
print("p-value:", p_value)
```

### e. Summarize his conclusion based on the P-value

As P-value is less than the significance level 0.05, we can reject $H_0$ that average shirt price of NuCloth is equal to that of supermarket's. That means there is enough statistical evidence to say that average shirt price of NuCloth is indeed greater than average shirt price of Supermarket.
